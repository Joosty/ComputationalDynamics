# Momentum: Linear and Angular

## Introduction

This chapter explores the concepts of linear and angular velocities for both particles and rigid bodies. These fundamental principles:
- Merge the concepts covered up until this point. 
- Underpin Newton's second law of motion, which is key to deriving equations that describe the motion of engineering systems using the particle and rigid body approximations.

## Linear Momentum

Linear momentum is a vector quantity defined as the product of an object's mass and velocity. The letter $\mathbf{L}$ is used to represent this quantity - typed in boldface to represent it as a vector.

### Particles

The linear momentum of a single particle (a point with mass) is defined as the product of its mass and velocity. Since the velocity of a particle is defined relative to a specific reference frame, the linear momentum is also defined relative to that frame.

### Notation

The linear momentum of a particle, $P$, in frame $N$ can be displayed as: ${}^{N}\mathbf{L}^{P}$

The mathematical definition of this is:

$${}^{N}\mathbf{L}^{P} = m{}^{N}\mathbf{v}^{P}$$

Where:

- $m$ is the mass of the particle.
- ${}^{N}\mathbf{v}^{P}$ is the velocity vector of the particle.

By virtue of mass centre definition, this reduces to a much simpler equation.

$${}^{N}\mathbf{L}^{S} = m^{N}\mathbf{v}^{S^{*}}$$


Where:

- $m$ is the mass of the system given by $\sum_{i} m_{i}$. 
- $\mathbf{v}^{S^{*}}$ is the velocity of the system's mass centre $S^{*}$.

## Angular Momentum

Angular momentum is also a vector quantity. It is quite generally defined as the moment of the linear momentum vector. The letter $\mathbf{H}$ is used to represent this quantity - typed in boldface to represent it as a vector.

### Particles

The angular momentum of the particle $P$ with respect to the point $O$ in frame $N$ is given by: ${}^{N}\mathbf{H}^{P/O}$

It is quite simply the moment of about $O$ of the linear momentum of $P$ in $N$.

The mathematical definition of this is:

$${}^{N}\mathbf{H}^{P/O} \triangleq \mathbf{r}^{OP} \times m{}^{N}\mathbf{v}^{P}$$

This is visualised in the figure below, showing the position vector $\mathbf{r}^{OP}$ from point $O$ to particle $P$ as well as the velocity vector of the particle ${}^{N}\mathbf{v}^{P}$. This is all in reference to the reference frame $N$.

### Systems of Particles

The angular momentum of a system of particles is quite simply the sum of the angular momentum vectors of each particle in the system. 

The angular momentum of a system $s$ of $n$ particles defined around the a point $O$ is given by: ${}^{N}\mathbf{H}^{P/O}$

The mathematical definition of this is:

$${}^{N}\mathbf{H}^{P/O} \triangleq \sum_{i} {}^{N}\mathbf{H}^{P_i/O} = \sum_{i}\mathbf{r}^{OP_i} \times m_i{}^{N}\mathbf{v}^{P_i}$$

```{attention}
Evaluating this integral can be quite challenging, so is there a simpler way to express the angular momentum of a body? Fortunately, yes! Let's explore how we can achieve this$\ldots$
```

First, we make use of the two-point theorem to rewrite the velocity of $P$ in terms of the mass centre of the rigid body. This is completed in two steps, resulting in equation $10.1$. 

$${}^{N}\mathbf{H}^{B/O} = \int_m (\mathbf{r}^{OP} \times ({}^{N}\mathbf{v}^{B^*}+{}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r}))\mathrm{d}m$$

$${}^{N}\mathbf{H}^{B/O} = \int_m (\mathbf{r}^{OP} \times {}^{N}\mathbf{v}^{B^*})\mathrm{d}m+\int_m(\mathbf{r}^{OP}\times({}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r}))\mathrm{d}m\tag{10.1}$$

So, we can now see that the angular momentum is the vector sum of two integrals. We will now examine the implication of each of these terms. 

We start with the first term which is given by:

$$\int_m (\mathbf{r}^{OP} \times {}^{N}\mathbf{v}^{B^*})\mathrm{d}m$$

This term can be dealt with as follows:

$$\int_m (\mathbf{r}^{OP} \times {}^{N}\mathbf{v}^{B^*})\mathrm{d}m = \int_m (\mathbf{r}^{OP} \times {}^{N}\mathbf{v}^{B^*})\mathrm{d}m$$
$$=\int_m (\mathbf{r}^{OP})\mathrm{d}m \times {}^{N}\mathbf{v}^{B^*}$$
$$=\mathbf{r}^{*}m \times {}^{N}\mathbf{v}^{B^*}$$
$$=\mathbf{r}^{*} \times m{}^{N}\mathbf{v}^{B^*}$$

Now, we can quite easily observe that this cross product is, in fact, the angular momentum of the mass centre $B^*$ about the point $O$ in the frame $N$. In other words,

$$\int_m (\mathbf{r}^{OP} \times {}^{N}\mathbf{v}^{B^*})\mathrm{d}m = {}^{N}\mathbf{H}^{B^*/O}\tag{10.2}$$

Next we can examine the second term, given by:

$$\int_m(\mathbf{r}^{OP}\times({}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r}))\mathrm{d}m$$

```{attention}
Prior to dealing with this term, note:

$$\int_m{\mathbf{\vec{r}}\mathrm{d}m} = 0$$

<div style="text-align: center;">$\mathbf{\vec{r}}$ is $\mathbf{r}^{OP}$</div>
```

Therefore,

$$\int_m(\mathbf{r}^{OP}\times({}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r}))\mathrm{d}m = \int_m(\mathbf{r}^{*}+\mathbf{r})\times({}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r})\mathrm{d}m$$ 

$$=\int_m(\mathbf{r}^{*}\times({}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r}))\mathrm{d}m + \int_m(\mathbf{r}\times({}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r}))\mathrm{d}m$$

$$=\mathbf{r}^{*}\times({}^{N}\boldsymbol{\omega}^{B}\times\int_m\mathbf{r}\mathrm{d}m) + \int_m(\mathbf{r}\times({}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r}))\mathrm{d}m$$

Since $\int_m{\vec{r}\mathrm{d}m} = 0$, the first term can be set to zero.

$$\int_m(\mathbf{r}^{OP}\times({}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r}))\mathrm{d}m=\cancel{\mathbf{r}^{*}\times({}^{N}\boldsymbol{\omega}^{B}\times\int_m\mathbf{r}\mathrm{d}m)} + \int_m(\mathbf{r}\times({}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r}))\mathrm{d}m$$

By definition of a mass centre:

$$\int_m(\mathbf{r}^{OP}\times({}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r}))\mathrm{d}m = \int_m(\mathbf{r}\times({}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r}))\mathrm{d}m$$

Now, if we examine the above cross product on the right-hand side and compare it to the definition of angular momentum of a rigid body, we can see that this is the angular momentum of the body $B$ about the mass centre $B^{*}$. In other words:

$$\int_m(\mathbf{r}\times({}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r}))\mathrm{d}m = {}^{N}\mathbf{H}^{B/B^*}$$

${}^{N}\mathbf{H}^{B/B^*}$ is frequently called the central angular momentum; central refers to the calculation of angular momentum about the mass centre. Further, the central angular momentum is easily calculated by the following formula:

$${}^{N}\mathbf{H}^{B/B^*} = [I]^{B/B^*}\cdot{}^{N}\boldsymbol{\omega}^{B}$$

Where:

- $[I]^{B/B^*}$ is the inertia matrix of $B$ about $B^*$; it is also called the __central inertia matrix__.
- ${}^{N}\boldsymbol{\omega}^{B}$ is the angular velocity of $B$ in $N$.

This says that the central angular momentum is given by the product of the inertia matrix of $B$ about $B^*$ and the angular velocity of $B$ in $N$; this proof is not shown here. In summary, we can see that this second integral eventually yields:

$$\int_m(\mathbf{r}\times({}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r}))\mathrm{d}m = [I]^{B/B^*}\cdot{}^{N}\boldsymbol{\omega}^{B} = {}^{N}\mathbf{H}^{B/B^*}\tag{10.3}$$

```{attention}
Indeed, one must be careful in ensuring that the the inertia matrix and the angular velocity were expressed in the same reference frames BEFORE taking the product.
```

Finally, we now return to Equation $10.1$ to write it in its most concise form as:

$${}^{N}\mathbf{H}^{B/O} = {}^{N}\mathbf{H}^{B/B^*} + {}^{N}\mathbf{H}^{B^*/O}\tag{10.4}$$

Equation $10.4$ is called the __transfer theorem for angular momentum__. This is because it allows us to calculate the angular momentum of a body about any point external to the body by transferring to its centre of mass and then to the point of interest. 

<!-- Coffee Break, 16/08/24 -->

### Special case

Say that we want to evaluate $^{N}H^{B/Q}$, where $Q$ is an arbitrary point rigidly fixed on the rigid body. $Q$ is not an elementary particle of the body. This is given and visualised as follows:

$${}^{N}\mathbf{H}^{B/Q} \triangleq \int_m \mathbf{r}^{QP} \times \mathrm{d}m({}^{N}\mathbf{v}^{P})$$
 
```{figure} ./images/28_cut.png
---
name: 28
---
```

Again, we make use of the two-point theorem to rewrite the velocity of $P$ in terms of $Q$. So, the
expression for angular momentum above is now:

$${}^{N}\mathbf{H}^{B/Q} = \int_m (\mathbf{r}^{QP}\times({}^{N}\mathbf{v}^{Q}+{}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r}^{QP}))\mathrm{d}m$$

Which then becomes:

$${}^{N}\mathbf{H}^{B/Q} = \int_m (\mathbf{r}^{QP}\times{}^{N}\mathbf{v}^{Q})\mathrm{d}m+ \int_m (\mathbf{r}^{QP}\times({}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r}^{QP}))\mathrm{d}m$$

$${}^{N}\mathbf{H}^{B/Q} = \left[\int_m (\mathbf{r}^{QP}\mathrm{d}m)\times{}^{N}\mathbf{v}^{Q}\right]+ \int_m (\mathbf{r}^{QP}\times({}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r}^{QP}))\mathrm{d}m$$

$${}^{N}\mathbf{H}^{B/Q} = \left[\mathbf{r}^{*}m\times{}^{N}\mathbf{v}^{Q}\right]+ \int_m (\mathbf{r}^{QP}\times({}^{N}\boldsymbol{\omega}^{B}\times\mathbf{r}^{QP}))\mathrm{d}m$$

The second term in the final expression, which is the integral, simplifies to a more straightforward result. Specifically, it is the product of the inertia matrix of $B$ about $Q$ and the angular velocity of $B$ in $N$. Although the proof of this simplification is not included here, it can be derived with relative ease. Thus, the equation above can be written more concisely as follows:

$${}^{N}\mathbf{H}^{B/Q} = \left[\mathbf{r}^{*}m\times{}^{N}\mathbf{v}^{Q}\right]+[I]^{B/Q}\cdot{}^{N}\boldsymbol{\omega}^{B}\tag{10.5}$$

Two key observations may now be made to simplify equation $10.5$:

__Observation 1__: If $Q$ is coincident with the mass centre $B^*$, then $\mathbf{r}^{*} = 0$. Thus:

$${}^{N}\mathbf{H}^{B/Q} = [I]^{B/Q}\cdot{}^{N}\boldsymbol{\omega}^{B}$$

__Observation 2__: If $Q$ is not the mass centre, but is a point that is fixed in both $B$ and $N$, then we have the following circumstance where:
 
$${}^{N}\mathbf{v}^{Q}=0$$

In this case, we end up with the same result as case 1:

$${}^{N}\mathbf{H}^{B/Q} = [I]^{B/Q}\cdot{}^{N}\boldsymbol{\omega}^{B}$$
